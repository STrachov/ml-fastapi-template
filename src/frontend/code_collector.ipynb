```python
import os
import glob

def collect_content_to_single_file(source_directory, output_file, exclude_dirs=None):
    if exclude_dirs is None:
        exclude_dirs = []  # Default to no exclusions
    
    with open(output_file, 'w', encoding='utf-8') as outfile:
        for file_path in glob.glob(os.path.join(source_directory, '**', '*.*'), recursive=True):
            # Check if the file is in an excluded directory
            if any(excluded in file_path for excluded in exclude_dirs):
                continue
            
            if os.path.isfile(file_path):  # Ensure it's a file
                with open(file_path, 'r', encoding='latin-1') as infile:
                    header_str = f'\n# {str(file_path)}\n'
                    outfile.write(header_str)
                    outfile.write(infile.read())
                    outfile.write('\n')  # Add a newline between files for clarity

dir_names = ['inner_pages']
exclude_dirs = ['__pycache__']  # List of folders to exclude

for dir_name in dir_names:
    print(dir_name)
    source_directory = f'./{dir_name}'
    output_file = f'collected_content_{dir_name}.txt'
    collect_content_to_single_file(source_directory, output_file, exclude_dirs)
```
